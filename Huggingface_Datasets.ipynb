## Hub'dan bir veri kümesi yükleyin

Hugging Face Hub'da mevcut bir veri kümesini keşfetmek ve veri kümesini indirmek ve oluşturmak için 🤗 Datasets kullanılır.

Bir veri setini indirmek için zaman ayırmadan önce, genellikle bir veri seti hakkında genel bilgileri hızlıca edinmek faydalıdır. Bir veri setinin bilgileri DatasetInfo içinde saklanır ve veri seti açıklaması, özellikler ve veri seti boyutu gibi bilgileri içerebilir.

Bir veri kümesi oluşturucusunu yüklemek ve veri kümesinin özniteliklerini indirmeye karar vermeden incelemek için load_dataset_builder() fonksiyonunu kullanın :

```python
from datasets import load_dataset
from datasets import load_dataset_builder
ds_builder = load_dataset_builder("cornell-movie-review-data/rotten_tomatoes")
```
